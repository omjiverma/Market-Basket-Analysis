# Market Basket Analysis

Understanding customer buying habits. By finding correlations and associations between different items that customers place in their ‘shopping basket,’ recurring patterns can be derived.

## Packages

```python
import pandas as pd
import numpy as np
from apyori import apriori
```

## Import Data

```python
df = pd.read_csv('grocery.csv')
df
```

## Formating Data to list

```python
tranx = []
for t in df.index:
    items = df.iloc[t].dropna()
    items=list(items.iloc[0:].values)
    tranx.append(items)
```

## Build the Apriori model

```python
rules = apriori(tranx, min_support=0.010, min_condifence=0.2, min_lift=3, min_length=2)
```

```python
result=[]
for x in rules:
    result.append(list(x.items))
```

## Items Frequently Bought Together

```python
result
```

## Outcome
We have successfully detected items that customers often buy together and we can put these items together in store for sales optimization
